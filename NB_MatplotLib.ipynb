# Introduction to Matplotlib

## 2 ways of creating Matplotlib
- pyplot()
- OO Method
- Matplotlib recommends the OOP Method

```python
# Import matplotlib and setup the figures to display within the notebooks
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
# A simple plot
plt.plot()
```

```python
plt.plot()
plt.show()
```

```python
# Adding some data
plt.plot([1,2,3,4])
plt.show()
```

```python
# Create some data
x = [1,2,3,4]
y = [11,22,33,44]
```

```python
plt.plot(x,y)
plt.show()
```

```python
# creating a plot with OO Method 
# Method 1 - add_subplot()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x,y)
plt.show()
```

```python
# creating a plot with OO Method
# Method 2 - add_axes()
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
# creating a plot with OO Method
# Method 3 - easier and more robust method 
fig,ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
# A matplotlib complete workflow
%matplotlib inline
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [11,22,33,44]

fig,ax = plt.subplots(figsize = (10,10))
ax.plot(x,y)


ax.set(title = 'Sample Simple Plot',xlabel = 'x-axis',ylabel = 'y-axis')

fig.savefig('simple-plot.png')
```

## Using Subplots to create mulitple plots on one figure

```python
# Option 1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(100))
plt.show()
```

```python
# Option 2

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,5))
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000))
plt.show()
```

## Plotting With pandas using the OO Method

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50
```

```python
over_50.plot(kind = 'scatter',x = 'age',y = 'chol',c = 'target',figsize = (10,5))
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = 'scatter', x = 'age', y = 'chol', c = 'target', ax = ax)
ax.set_xlim([50,100])
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,6))

scatter = ax.scatter(over_50['age'],over_50['chol'],c = over_50['target'])
ax.set(title = 'Heart Disease and Cholestrol Levels', xlabel = 'Age', ylabel = 'Cholesterol')
ax.legend(*scatter.legend_elements(), title = 'Target')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,6))
scatter = ax.scatter(over_50['age'],over_50['chol'],c = over_50['target'])

ax.set(title = 'Heart Disease and Cholesterol Levels', xlabel = 'Age', ylabel = 'Cholesterol')
ax.legend(*scatter.legend_elements(), title = 'Target')
ax.axhline(over_50['chol'].mean(), linestyle = '--')
plt.show()
```

### Adding subplots using OO Method

```python
fig, (ax0, ax1) = plt.subplots(nrows=2,ncols=1,sharex = True, figsize = (10,8))

scatter = ax0.scatter(over_50['age'],over_50['chol'],c=over_50['target'])
ax0.set(title='Heart Disease and Cholesterol Levels', ylabel = 'Cholesterol')
ax0.legend(*scatter.legend_elements(),title = 'Target')

ax0.axhline(y=over_50['chol'].mean(), color = 'b',linestyle = '--', label = 'Average')

scatter = ax1.scatter(over_50['age'], over_50['thalach'],c = over_50['target'])
ax1.set(title = 'Heart Disease and Max heart Rate Levels',xlabel = 'Age',ylabel = 'Max Hear rate')
ax1.legend(*scatter.legend_elements(),title = 'Target')
ax1.axhline(y = over_50['thalach'].mean(),color ='b',linestyle = '--',label = 'Average');
```
